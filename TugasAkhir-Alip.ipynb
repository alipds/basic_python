```python
# import library
import datetime
import random
import math
import os


print("Hai aku AIs, asisten virtual pertamamu")
print("Siapa namamu?")
# ganti dengan sebuah nama
nama  = input()
# variabel tanggal
x = datetime.datetime.now()
tanggal = x.strftime("%x")
# default variabel untuk pertanyaan tidak diketahui
default = "maaf, aku belum paham yang kamu tanyakan"


print("\nSalam kenal {}".format(nama))
print("Apakah ada yang bisa aku bantu?\n")

#munculkan menu
def menuUtama():
    print("""
    ---- MENU UTAMA ----
    1. Kalkulator
    2. Agenda
    3. Chatbot
    4. Keluar
    """)
    return

menuUtama()

#input untuk memilih menu
menu = int(input("Pilih Menu (1/2/3/4): "))

while True:
#---KALKULATOR
#buat fungsi untuk menyelesaikan beberapa persoaalan hitungan. variasi hitungan bebas
    if menu == 1:
        #penjumlahan
        def tambah(a, b):
            return a + b
        #pengurangan
        def kurang(a, b):
            return a - b
        #perkalian
        def kali(a, b):
            return a * b
        #pembagian
        def bagi(a, b):
            return a / b

        a = int(input("Angka pertama: "))
        c = input("Operator (+), (-), (*), (/): ")
        b = int(input("Angka kedua: "))

        while True:
            if c == "+":
                print(a, "+", b, "=", tambah(a, b))
            elif c == "-":
                print(a, "-", b, "=", kurang(a, b))
            elif c == "*":
                print(a, "*", b, "=", kali(a, b))
            elif c == "/":
                print(a, "/", b, "=", bagi(a, b))
            else:
                print("Operator salah.")
            d = input("Lanjut / Kembali: ")
            if d == "Lanjut":
                a = int(input("Angka pertama: "))
                c = input("Operator: ")
                b = int(input("Angka kedua: "))
            elif d == "Kembali":
                break
        menuUtama()
        menu = int(input("Pilih Menu (1/2/3/4): "))

#---AGENDA
#Buat fungsi untuk mencatat kegiatan harian yang di simpan kedalam sebuah file.
    if menu == 2:
        def menuAgenda():
            print("""
            --- Menu Agenda ---
            1. Tulis Agenda
            2. Baca Agenda
            3. Tambah Agenda
            4. Kembali
              """)
            return
    
        menuAgenda()

        menu_agenda = int(input("Pilih Menu Agenda: "))

        while True:
            if menu_agenda == 1:
                kegiatan = input("Tulis Kegiatan: ")
                teks = "Tanggal: {}\n{}".format(tanggal, kegiatan)
                f_agenda = open("Agenda.txt", "w")
                f_agenda.write(teks)
                f_agenda.close()
            elif menu_agenda == 2:
                f_agenda = open("Agenda.txt", "r")
                if f_agenda.mode == "r":
                    agenda = f_agenda.read()
                    print(agenda)
            elif menu_agenda == 3:
                kegiatan = input("Tulis Kegiatan: ")
                teks = "\n{}".format(kegiatan)
                f_agenda = open("Agenda.txt", "a+")
                f_agenda.write(teks)
                f_agenda.close()
            elif menu_agenda == 4:
                break
            else:
                print("Maaf Menu Agenda Tidak Tersedia")
            menuAgenda()
            menu_agenda = int(input("Pilih Menu Agenda: "))
        menuUtama()
        menu = int(input("Pilih Menu (1/2/3/4): "))

#---BOOTHCHAT
# Membuat objek dictionary berisi berbagai opsi jawaban
    if menu == 3:
        # list jawaban untuk pertanyaan tentang tanggal
        jawaban_nama = [
            "nama saya  {0}".format(nama),
            "orang-orang memanggil saya {0}".format(nama),
            "panggil saja saya {0}".format(nama)
            ]

        # list jawaban untuk pertanyaan tentang tanggal
        jawaban_tanggal = [
            "hari ini tanggal {0}".format(tanggal),
            "ya ampun masa tidak tahu, hari ini tanggal {0}".format(tanggal)
            ]

        # opsi pertanyaan yang bisa dijawab
        pertanyaan = {
            "nama kamu siapa?": jawaban_nama,
            "kamu siapa?" : jawaban_nama,
            "tanggal berapa hari ini?": jawaban_tanggal,
            "hari ini tanggal berapa?" : jawaban_tanggal,
            "default": default
            }

        # list jawaban untuk sebuah argument selain pertanyaan
        statement =  [
            'ceritakan lebih banyak!',
            'kenapa kamu berpikir begitu?',
            'sudah berapa lama kamu merasa seperti ini?',
            'Itu sangat menarik!',
            'oh wow!',
            ':)'
            ]

        # respon keseluruhan
        responses = {
            'pertanyaan' : pertanyaan,
            'statement' : statement
            }

        print("""
        Haloo... Selamat Datang di Chatbot!

        *ketik 'selesai' untuk keluar
        """)

        response = input("")
        
        while True:
            if response in responses['pertanyaan']:
                if "siapa" in response:
                    jawabNama = random.choice(jawaban_nama)
                    print(jawabNama)
                elif "tanggal" in response:
                    jawabTanggal = random.choice(jawaban_tanggal)
                    print(jawabTanggal)
            elif response == "selesai":
                break
            else:   
                jawabStatement = random.choice(statement)
                print(jawabStatement)
            response = input()

        print("Terimakasih.")
        menuUtama()
        menu = int(input("Pilih Menu (1/2/3/4): "))
        
    if menu == 4:
        break
        
print("Terimakasih, senang bisa membantu.")
```
